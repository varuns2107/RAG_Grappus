```python
## create a simple txt file

import os
os.makedirs("../data/text_files", exist_ok=True)
```

```python
sample_texts= {
    "../data/text_files/python_intro.txt":"""Python Programming Introduction

Python is a high-level, interpreted programming language known for its simplicity and readability.
Created by Guido van Rossum and first released in 1991, Python has become one of the most popular
programming languages in the world.

Key Features:
- Easy to learn and use
- Extensive standard library
- Cross-platform compatibility
- Strong community support

Python is widely used in web development, data science, artificial intelligence, and automation.""",
    
    "../data/text_files/machine_learning.txt": """Machine Learning Basics

Machine learning is a subset of artificial intelligence that enables systems to learn and improve
from experience without being explicitly programmed. It focuses on developing computer programs
that can access data and use it to learn for themselves.

Types of Machine Learning:
1. Supervised Learning: Learning with labeled data
2. Unsupervised Learning: Finding patterns in unlabeled data
3. Reinforcement Learning: Learning through rewards and penalties

Applications include image recognition, speech processing, and recommendation systems"""
}

for filepath,content in sample_texts.items():            
    with open(filepath, 'w', encoding='utf-8') as fnc:
        fnc.write(content) 
        
print ("We have created the sample file successfully !!!")

#File path is nothing but the name of my file where i am storing this data.
#Here our filepath is ""../data/text_files/python_intro.txt" as we are storing in data in text_files as the name of python_intro.txt
```

```python
### Reading textfile created using Textloader

from langchain_community.document_loaders import TextLoader         ##Textloader for loading text.
loader = TextLoader("../data/text_files/python_intro.txt", encoding="utf-8")
document = loader.load()
print(document)
```

```python
from langchain_community.document_loaders import DirectoryLoader
dir_loader= DirectoryLoader(
    "../data/text_files",
    glob="**/*.txt",    ## This is the pattern to match files with i.e. .txt and load all .txt files (** means search all subfolder and *.txt = all txt files)
    
    loader_cls=TextLoader,    ## Class of the loader, here it is text so textloader if it was pdf then PypdfLoader  would be used
    
    loader_kwargs={'encoding' : 'utf-8'},       ##Loader keyword arguments, i.e extra arguments or filters or extra settings/options for loader
    
    show_progress=True      ##To show real time progress eg. Loading documents:  45%|█████████         | 450/1000
)
document = dir_loader.load()
document
```

```python
from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader, PyMuPDFLoader
pdf_loader= DirectoryLoader(
    "../data/pdf_files",
    
    glob="**/*.pdf",    ## This is the pattern to match files with i.e. .pdf and load all .pdf files (** means search all subfolder and *.pdf = all pdf files)
    
    loader_cls=PyMuPDFLoader,   ## Class of the loader, here it is pdf so PyMuPDFLoader is used
    
    show_progress=True      ##To show real time progress eg. Loading documents:  45%|█████████         | 450/1000
)
document = pdf_loader.load()
document
```
